### <div style="direction: rtl; text-align: center;"> خواندن فایل متنی و جدا کردن بر اساس هر خط </div>

```python
with open('stop_words.txt') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [str(line).replace('\n', '') for line in stopwords]
print(stopwords)
```


### <div style="direction: rtl; text-align: center;"> خواندن فایل csv  </div>

```python
#dataset: https://gist.github.com/netj/8836201
import csv
with open('iris.csv') as csv_file:
    reader = csv.reader(csv_file, delimiter=",")
    count = 0
    for row in reader:
        if count == 0:
            print(f'Column name: {", ".join(row)}')
            count += 1
        else:
            print(f'Row is {row}')
            count += 1
    print('finished')
```
